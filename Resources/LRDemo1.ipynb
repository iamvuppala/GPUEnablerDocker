```octave
X=data(1:100,1:2);
Y=data(1:100,3);
```

```octave
size(X)
size(Y)
```

```octave
pos=find(Y == 1);
neg=find(Y == 0);
```

```octave
figure; hold on;
plot(X(pos, 1), X(pos, 2), 'k+','LineWidth', 2, 'MarkerSize', 7);
plot(X(neg, 1), X(neg, 2), 'ko', 'MarkerFaceColor', 'y','MarkerSize', 7);
hold off;
```
